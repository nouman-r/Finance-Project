# EDA

Let's explore the data a bit! 
** What is the max Close price for each bank's stock throughout the time period?**

```python
bank_stocks.xs('Close', level='Stock Info', axis=1).max().round(2)
```

** Create a new empty DataFrame called returns. This dataframe will contain the returns for each bank's stock. returns are typically defined by:**

$$r_t = \frac{p_t - p_{t-1}}{p_{t-1}} = \frac{p_t}{p_{t-1}} - 1$$

```python
returns = pd.DataFrame()
```

```python
#bank_stocks.xs('Close', level='Stock Info', axis=1)['BAC'].pct_change()
for ticker in tickers:
    returns[ticker+ ' Return'] = bank_stocks.xs(ticker, level=0, axis=1)['Close'].pct_change()
```

```python
returns.head()
```

```python
import seaborn as sns
sns.pairplot(returns)
```

## Moving Averages

Let's analyze the moving averages for these stocks in the year 2008. 

** Plot the rolling 30 day average against the Close Price for Bank Of America's stock for the year 2008**

```python
bank_stocks.loc['2008-01-01':'2008-12-31'].xs('Close', level='Stock Info', axis=1)['BAC'].rolling(30).mean().plot(figsize= (10,6), label = 'Rolling 30 day Avg')
bank_stocks.loc['2008-01-01':'2008-12-31'].xs('Close', level='Stock Info', axis=1)['BAC'].plot(label='BAC')
plt.legend()
```

** Create a heatmap of the correlation between the stocks Close Price.**

```python
sns.heatmap(bank_stocks.xs('Close', level='Stock Info', axis=1).corr(),annot=True, cmap='coolwarm')
```

** Optional: Use seaborn's clustermap to cluster the correlations together:**

```python
sns.clustermap(bank_stocks.xs('Close', level='Stock Info', axis=1).corr(),annot=True, cmap='RdYlBu')
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock Info').corr().iplot(kind='heatmap',colorscale='rdylbu')
```
